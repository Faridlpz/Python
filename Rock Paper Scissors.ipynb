```python
from tkinter import *
from random import choice



root = Tk()

root.geometry('300x220')

vector = ['tijera','papel','piedra']

    
myText=StringVar()

tijera ='Tijera'
papel = 'papel'
piedra ='piedra' 


# Definimos una función a ejecutar al clic el botón
def juegotijera():
    palabra_aleatoria = choice(vector)
    if palabra_aleatoria == 'tijera':
        res= 'tijera vs ',' La maquina escogio: ' ,palabra_aleatoria,'  EMPATE'
    elif palabra_aleatoria == 'papel':
        res= 'tijera vs ',' La maquina escogio:',palabra_aleatoria,'  GANASTE'
    elif palabra_aleatoria == 'piedra':
        res= 'tijera vs ',' La maquina escogio: ', palabra_aleatoria,'  PERDISTE'   
   
    return myText.set(res) 

def juegopapel():
    palabra_aleatoria = choice(vector)
    if palabra_aleatoria == 'tijera':
        res= 'Papel vs ',' La maquina escogio: ' ,palabra_aleatoria,'  PERDISTE'
    elif palabra_aleatoria == 'papel':
        res= 'Papel vs ',' La maquina escogio:',palabra_aleatoria,'  EMPATE'
    elif palabra_aleatoria == 'piedra':
        res= 'Papel vs ',' La maquina escogio: ', palabra_aleatoria,'  GANASTE'   
   
    return myText.set(res)   

def juegopiedra():
    palabra_aleatoria = choice(vector)
    if palabra_aleatoria == 'tijera':
        res= 'Piedra vs ',' La maquina escogio: ' ,palabra_aleatoria,'  GANASTE'
    elif palabra_aleatoria == 'papel':
        res= 'Piedra vs ',' La maquina escogio:',palabra_aleatoria,'  PERDISTE'
    elif palabra_aleatoria == 'piedra':
        res= 'Piedra vs ',' La maquina escogio: ', palabra_aleatoria,'  EMPATE'   
   
    return myText.set(res)   
    
    

resultado = Label(root,text="",textvariable=myText).pack()
     

# Enlezamos la función a la acción del botón
Button(root, text="Tijera", command=juegotijera).pack()
Button(root, text="Papel", command=juegopapel).pack()
Button(root, text="Piedra", command=juegopiedra).pack()






root.mainloop() 
```
